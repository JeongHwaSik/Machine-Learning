```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as so

#%matplotlib notebook
plt.close("all")

x1 = np.linspace(-4,4,100)
x2 = np.linspace(-2,2,100)
vx, vy = np.meshgrid(x1, x2)

def f(x):
    return x[0]**2 + x[0]*x[1] + 10*x[1]**2 - 5*x[0] - 3*x[1]

f1 = plt.figure()
ax1 = f1.add_subplot(projection="3d")
ax1.plot_surface(vx,vy,f([vx,vy]))

f2 = plt.figure()
ax2 = f2.add_subplot(111)
ax2.contourf(vx,vy,f([vx,vy]))
ax2.set(xlim=(-4, 4), ylim=(-2, 2))

# gradient descent
eta = 0.085
x0 = np.array([-3,-1])
h=np.sqrt(np.finfo(float).eps)
for i in range(20):
    # How can I connect the dots????
    ax2.plot(x0[0],x0[1],"ro")
    grad=so.approx_fprime(x0, f, h)
    x0 = x0 - eta*grad

# gradient descent with Momentum
eta = 0.085
alpha = 0.1
x0 = np.array([-3,-1])
h=np.sqrt(np.finfo(float).eps)
for i in range(20):
    d = np.array([0.0,0.0])
    ax2.plot(x0[0],x0[1],"bo")
    grad=so.approx_fprime(x0, f, h)
    d -= x0
    x0 = x0 - eta*grad
    d += x0
    x0 = x0 + alpha*d
    
plt.show()
```
